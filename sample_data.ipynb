in the following function we're trying to do the following
* check if there are any duplicated rows
* count the number of missing rows and display the same.
* drop the duplicates and display the action.
* finally return the dataframe with the new changes.


```python
def duplicated_data(data):
    
    if (data.duplicated(subset = None, keep = "first").any() == True):
        
        duplicated_count = data.duplicated().value_counts()
        
        num_duplicates = duplicated_count[True]
        print(f"{num_duplicates} duplicates found.")
        data.drop_duplicates(subset = None, keep = "first", inplace = True)
        print("Duplicates removed.")
            
        data.reset_index(drop = True, inplace = True)
            
        return data
    else:
        print("No duplicates found.")
        return data
    
sample_data = duplicated_data(sample_data)
```

```python
""" In general, we shall create a variable miss to store all the columns with missing data and 
sort them in descending order. A variable percenatge_miss will be created to check out the ratio of 
missing data to the len of the column. We shall also create a data frame "missing" to display
the "miss" and "percentage miss" data and finally drop all data with no missing value so that we can 
only remain with the data that has missing values in our data frame.
"""
def missing_values(data):
    miss = data.isnull().sum().sort_values(ascending = False)

    percentage_miss = (np.round((data.isnull().sum() / len(data)) * 100, 3)).sort_values(ascending = False)

    missing = pd.DataFrame({"Missing Values": miss, "Percentage(%)": percentage_miss})

    missing.drop(missing[missing["Percentage(%)"] == 0].index, inplace = True)

    return missing


missing_data = missing_values(sample_data)
missing_data
```

```python
"""
Here we shall drop all empty columns. ie. where the percentage of the missing
data is 100%
"""
def remove_columns_with_100_percent_missing(data):
    missing_data = missing_values(data)
    columns_to_remove = missing_data[missing_data["Percentage(%)"] == 100].index
    data.drop(columns = columns_to_remove, inplace = True)
    return data

sample_data = remove_columns_with_100_percent_missing(sample_data)
```

```python
missing_data = missing_values(sample_data)
missing_data
```

```python
"""
We shall concatenate the columns with dates to form a new column then convert
the same into the the "datetime format" and to be able to address the columns with missing 
dates, we shall employ the errors="coerce" so that python can not return the missing value
error.
"""
def datetime(data, new_col, day, month, year, date_format, index):
    
    data[new_col] = data[day].astype(str) + "-" + data[month] + "-" + data[year].astype(str)
    
    data[new_col] = pd.to_datetime(data[new_col], format=date_format, errors = "coerce")
    
    data.insert(index, new_col, data.pop(new_col))
    
    data.drop(columns=[day, month, year], inplace=True)
    
    return data
```

```python
sample_data = datetime(sample_data, "Today's Date", 'T-Day', 'T-Month', 'T-Year', "%d-%b-%Y", 0)
```
